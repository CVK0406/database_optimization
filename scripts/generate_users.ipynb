```python
import csv
import random
import uuid

MISSING_COUNT = 559

# Tỉ lệ thực tế các bang ở Brazil theo dataset Olist
STATES =['SP', 'RJ', 'MG', 'RS', 'PR', 'SC', 'BA', 'Outros']
STATE_WEIGHTS =[42.0, 13.0, 11.6, 5.5, 5.0, 3.6, 3.4, 15.9]

CITIES_BY_STATE = {
    'SP':['sao paulo', 'campinas', 'guarulhos', 'sao bernardo do campo'],
    'RJ': ['rio de janeiro', 'niteroi', 'nova iguacu'],
    'MG':['belo horizonte', 'juiz de fora', 'contagem'],
    'RS': ['porto alegre', 'caxias do sul'],
    'PR': ['curitiba', 'londrina'],
    'SC':['florianopolis', 'joinville'],
    'BA': ['salvador', 'feira de santana'],
    'Outros':['brasilia', 'fortaleza', 'manaus']
}

print(f"Đang sinh {MISSING_COUNT} Users theo phân phối chuẩn...")

with open('missing_users.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['customer_id', 'customer_unique_id', 'zip_code', 'city', 'state'])

    for _ in range(MISSING_COUNT):
        customer_id = uuid.uuid4().hex
        customer_unique_id = uuid.uuid4().hex

        # Random State dựa trên trọng số phần trăm
        state = random.choices(STATES, weights=STATE_WEIGHTS, k=1)[0]
        # Chọn ngẫu nhiên 1 thành phố thuộc bang đó
        city = random.choice(CITIES_BY_STATE[state])

        zip_code = str(random.randint(10000, 99999))
        writer.writerow([customer_id, customer_unique_id, zip_code, city, state])

print("Hoàn tất sinh Users!")
```
